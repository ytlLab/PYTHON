# 語音文字處理

## gTTS：文字轉語音

```python
!pip install gTTS
```

```python
import gtts
import IPython.display as display
```

```python
txt = 'gTTS可以透過線上翻譯，將文字轉換為語音，並將語音存檔'
tts = gtts.gTTS(text=txt, lang='zh-TW')                     #text代表要轉換的文字字串 lang為語言代碼
tts.save('gtts.mp3')                                        #將轉換的語音儲存為mp3檔
display.Audio("gtts.mp3", autoplay=False)
```

```python
f = open("gtts2.mp3", "wb") #如果要轉換的文字太長，可以將文字分解為多個段落，分別轉換為語音後，共同存入一個同一個語音檔
tts1 = gtts.gTTS(text='gTTS可以透過線上翻譯，將文字轉換為語音，並將語音存檔', lang='zh-TW')
tts1.write_to_fp(f)
tts2 = gtts.gTTS(text='將文字分解為多個段落，分別轉換為語音', lang='zh-TW')
tts2.write_to_fp(f)
display.Audio("gtts2.mp3", autoplay=True)
```

## 微軟的 edge-tts

```python
!pip install edge-tts
```

```python
import edge_tts
import asyncio
from IPython.display import Audio

text = "這段語音聽起來比一般的合成音更自然。"
voice = "zh-TW-HsiaoChenNeural" # 台灣女聲
output_file = "edge_output.mp3"

async def main():
    communicate = edge_tts.Communicate(text, voice)
    await communicate.save(output_file)

await main()

Audio(output_file)
```

```python
!edge-tts --list-voices
```

```python
import asyncio
import edge_tts

async def show_voices():
    voices = await edge_tts.VoicesManager.create()
    # 尋找包含 'zh' (中文) 的聲音
    chinese_voices = voices.find(Language="zh")
    for v in chinese_voices:
        print(f"名稱: {v['ShortName']}, 性別: {v['Gender']}, 地區: {v['Locale']}")

await show_voices()
```

```python
import whisper
from IPython.display import Audio

# 2. 載入模型 (提供 tiny, base, small, medium, large-v3 五種大小)
# 'base' 速度快且準確度足夠，'medium' 以上對中文更好
model = whisper.load_model("base")

# 3. 進行轉錄
# 假設你有一個檔案叫做 "audio.mp3"
result = model.transcribe("record1.wav", fp16=False)

# 4. 印出文字
print("-" * 30)
print(result["text"])
```

## 文字翻譯 DeepLX

```python
!pip install deep-translator
```

```python
from deep_translator import GoogleTranslator

# 使用 GoogleTranslator 介面（這是目前最穩定的第三方封裝）
text = "Hello, how are you doing today?"
translated = GoogleTranslator(source='auto', target='zh-TW').translate(text)

print(f"翻譯結果: {translated}")
```

```python
from deep_translator import GoogleTranslator

texts = ["Good morning", "Have a nice day", "See you later"]
translator = GoogleTranslator(source='en', target='zh-TW')

# 批次翻譯
translations = translator.translate_batch(texts)

for original, translated in zip(texts, translations):
    print(f"{original} -> {translated}")
```
