完美的數：將一正整數的所有公因數加總，總和與此正整數相等，稱此數為完美的數(perfect number)。例如，6是第一個完美的數因為6=3+2+1，下一個是28=14+7+4+2+1。
請寫一程式找出10000以下，四個完美的數為：6 28 496 8128


```python
#練習
n = int(input("請輸入大樓的樓層數："))
print("本大樓具有的樓層為：")
if(n > 3):
	n += 1
for i in range(1, n+1):
	if(i==4):
    		continue
  print(i, end=" ")
print()
```

```python
#練習
n = int(input("請輸入大於 1 的整數："))
if(n == 2):
	print("2 是質數！")
else:
  for i in range(2, n):
    if(n % i == 0):
      print("%d 不是質數！" % n)
      break
  else:
    print("%d 是質數！" % n)
```

```python
#練習
a = ['a','b','c','d','e']
a.insert(3,100)
b = ['a','b','c','d','e']
b.insert(0,100)
c = ['a','b','c','d','e']
c.insert(100,100)
d = ['a','b','c','d','e']
d.insert(-1,100)
print(a)   # ['a', 'b', 'c', 100, 'd', 'e']
print(b)   # [100, 'a', 'b', 'c', 'd', 'e']
print(c)   # ['a', 'b', 'c', 'd', 'e', 100]
print(d)   # ['a', 'b', 'c', 'd', 100, 'e']
```
