### 類別的建構式
初始化方法有一個固定名稱是”__init__( )”,，

寫法是init左右皆是2個底線字元，init其實是initialization的縮寫，

通常又將這類初始化的方法稱建構方法(constructor)

建構式必須使用def __init__(self, 參數1,參數2,...)，

參數self是必須的，同時需要放在最前面，代表建立的物件，

其他參數可選擇。如此在類別中就可以使用self.屬性、self.方法執行類別的屬性和方法。

```python
#建立Animal類別，並建立__init__()建構式與sing方法
class Animal():  	#定義類別
  def __init__(self, name):
    self.na = name  #定義屬性

  def sing(self):   	#定義方法
    print(self.na + "，很會唱歌!")

bird = Animal("鸚鵡")  #以 Animal 類別，建立一個名叫鸚鵡的 bird物件
#bird.na="abc"
print(bird.na) #鸚鵡
bird.sing()  	#鸚鵡，很會唱歌!
```

```python
#屬性初始值的設定
class Animal():  	#定義類別
  def __init__(self, name,age):
    self.na = name  #定義屬性
    self.ag = age

  def sing(self):   	#定義方法
    print(self.na + str(self.ag) + "歲，很會唱歌!")

  def grow(self,year):  #定義方法
    self.ag += year

bird = Animal("鸚鵡",1) #以 Animal 類別，建立一個名叫鸚鵡、1歲大的 bird物件
bird.grow(1) 	#長大1歲
#bird.ag=5
bird.sing()  	#鸚鵡2歲，很會唱歌!

```

類別封裝

Python提供私用(private)屬性與私用方法，只有類別內部可以使用，類別外部無法使用，此種觀念稱為封裝。

在屬性和方法前加上__，就成為私用屬性和方法。例如__name、__age、__sing

```python
#類別封裝
class Animal():  	#定義類別
  def __init__(self, name,age):
    self.__name = name  #定義私用屬性
    self.__age = age

  def __sing(self):   	#定義私用方法
    print(self.__name + str(self.__age),end= "歲，很會唱歌，")

  def talk(self):    	#定義共用方法
    self.__sing()  	#使用私用方法
    print("也會模仿人類說話!")

bird = Animal("灰鸚鵡",2) #以 Animal 類別，建立一個名叫灰鸚鵡、2歲大的 bird物件
bird.talk()   	#灰鸚鵡2歲，很會唱歌，也會模仿人類說話!
#bird.__name="abc"
#bird.__age = -1  #設定無效
bird.talk()  	#灰鸚鵡2歲，很會唱歌，也會模仿人類說話!
#bird.__sing()   #執行出現錯誤
```

類別繼承建立子類別的語法：

class 類別名稱(父類別):

例如：建立類別Bird繼承Animal類別，

Animal是父類別，Bird是子類別
class  Bird(Animal)

```python
#類別繼承
class Animal():  	#定義父類別
  def __init__(self, name):
   self.name = name  #定義共用屬性

  def fly(self):    	#定義共用方法
    print(self.name + "很會飛!")

class Bird(Animal):  	#定義子類別
  def __init__(self, name):
    self.name = "粉紅色" + name  #覆寫父類別的建構式

  def sing(self):   	#定義子類別的方法
    print(self.name + "也愛唱歌!")

pigeon = Animal("小白鴿")#以 Animal 類別，建立一個名叫小白鴿的 pigeon 物件
pigeon.fly()  #小白鴿很會飛!
#pigeon.sing()

parrot = Bird("小鸚鵡")  #以 Bird 類別，建立一個名叫小鸚鵡的 parrot 物件
parrot.fly()  #粉紅色小鸚鵡很會飛!
parrot.sing() #粉紅色小鸚鵡也愛唱歌!
```

類別繼承

若子類別與父類別擁有相同的屬性和方法，子類別會先尋找自己是否有此名稱或方法，
若沒有才使用父類別的屬性和方法

若要使用父類別的方法，則需要使用super()方法操作。

```python
#類別繼承
class Animal():  	#定義父類別
  def __init__(self,name):
    self.name = name  #定義共用屬性

  def fly(self):    	#定義共用方法
    print(self.name + "很會飛!")

class Bird(Animal):  	#定義子類別
  def __init__(self,name,age):
   super().__init__(name) #執行父類別的 __init__()方法
   self.age = age   #定義子類別共用屬性

  def fly(self):   	#定義子類別共用方法
    print(str(self.age),end="歲")
    super().fly() #執行父類別的 fly方法

pigeon = Animal("小白鴿") #以 Animal 類別，建立一個名叫小白鴿的 pigeon 物件
pigeon.fly()  #小白鴿很會飛!

parrot = Bird("小鸚鵡",2) #以 Bird 類別，建立一個名叫小鸚鵡、2歲大的 parrot 物件
parrot.fly()  #2歲小鸚鵡很會飛!
```

多型

上個例子不同類別有相同的方法名稱，此稱為多型，但其實多型不一定需要有繼承關係，可以使用同一個方法名稱卻產生不同的功能。


```python
#多型
class Animal():      	#定義父類別
  def fly(self):   	#定義共用方法
    print("時速 20公里!")
class Bird(Animal):  	#定義子類別
  def fly(self,speed): #定義共用方法
    print("時速 " + str(speed) + "公里!")

class Plane():       	#定義類別
  def fly(self):   	#定義共用方法
    print("時速 1000公里!")

def fly(speed):      	#定義函式
  print("時速 " + str(speed) + "英哩!")

animal = Animal() #以 Animal 類別建立 animal 物件
animal.fly()  #時速 20公里!

bird = Bird() #以 Bird 類別建立 bird 物件
bird.fly(60)  #時速 60公里!

plane=Plane() #以 Plane 類別建立 plane 物件
plane.fly()   #時速 1000公里!

fly(5)    	#時速 5英哩!
```

多重繼承

語法：class 類別名稱(父類別1,父類別2,...父類別n):
如果父類別有相同的屬性或方法，會由左至右搜尋。


```python
#多重繼承
class Father():     	#定義父類別
  def say(self):  	#定義共用方法
    print("明天會更好!")
class Mother():     	#定義父類別
  def say(self):  	#定義共用方法
    print("包容、尊重!")
class Child(Father,Mother): #定義子類別
	pass

child = Child() #建立 child 物件
child.say() 	#明天會更好!
```
