```python
filename = 'obama.jpg'
#filename = 'person1.jpg'
#filename = 'person2.jpg'
#filename = 'person3.jpg'

image = face_recognition.load_image_file(filename)
landmarks = face_recognition.face_landmarks(image)
print(landmarks)
print('鼻樑位置：{}'.format(landmarks[0]['nose_bridge']))

img = Image.open(filename)
drawing = ImageDraw.Draw(img)
for landmark in landmarks:
    for feature in landmark.keys():
        drawing.line(landmark[feature], width=5)
plt.imshow(img)
plt.show()
```

```python
img1 = face_recognition.load_image_file('sample1.jpg')
img2 = face_recognition.load_image_file('sample2.jpg')
img3 = face_recognition.load_image_file('sample3.jpg')
encoding1 = face_recognition.face_encodings(img1)[0]
encoding2 = face_recognition.face_encodings(img2)[0]
encoding3 = face_recognition.face_encodings(img3)[0]
known_faces = [encoding1, encoding2, encoding3]
names = ['jeng', 'chiou', 'david']

unknown = face_recognition.load_image_file("catch.jpg")
#unknown = face_recognition.load_image_file("lily2.jpg")

encoding_unknown = face_recognition.face_encodings(unknown)[0]
results = face_recognition.compare_faces(known_faces, encoding_unknown)
print(results)
face = ''
for i in range(len(results)):
  if results[i]: face = face + names[i] + '  '
if face == '': print('圖片中辨識不到資料庫中人臉！')
else: print('圖片中的人臉：' + face)
```

### 人臉驗證

```python
from deepface import DeepFace

face1 = 'bear1.jpg'
face2 = 'bear2.jpg'
#face2 = 'jeng1.jpg'
#face2 = 'david1.jpg'

# 使用 'VGG-Face' 或 'Facenet512'，這些模型與新版 TensorFlow 完全相容
model_name = "VGG-Face"

try:
    # 簡化呼叫方式，DeepFace 會自動幫你處理模型下載與建立
    result = DeepFace.verify(
        img1_path = face1,
        img2_path = face2,
        model_name = model_name,
        enforce_detection = False
    )

    print(result)

    if result["verified"]:
        print(f'使用 {model_name} 驗證：兩張圖片是同一人！')
    else:
        print(f'使用 {model_name} 驗證：兩張圖片不是同一人！')

except Exception as e:
    print(f"發生錯誤: {e}")
```

```python
from deepface import DeepFace

face1 = 'bear1.jpg'
face2 = 'bear2.jpg'
#face2 = 'jeng1.jpg'

# 移除掉會造成錯誤的 "DeepFace" 模型
models = ["VGG-Face", "Facenet", "OpenFace", "DeepID", "Dlib", "ArcFace"]

result = []

for model in models:
    try:
        print(f"正在使用 {model} 進行驗證...")
        ret = DeepFace.verify(face1, face2, model_name=model, enforce_detection=False)
        # 為了讓結果更易讀，我們可以把模型名稱加入結果中
        ret['model_used'] = model
        result.append(ret)
    except Exception as e:
        print(f"模型 {model} 執行失敗: {e}")

# 列印最終所有模型的比對結果
print("\n--- 最終比對報告 ---")
for r in result:
    print(f"模型: {r['model_used']} | 是否為同一人: {r['verified']} | 距離: {r['distance']:.4f}")
```

### 搜尋人臉

```python
import pandas as pd
import numpy as np
```

```python
face1 = 'bear2.jpg'
#face1 = 'obama.jpg'
#face1 = 'person1.jpg'

# 執行搜尋
results = DeepFace.find(img_path=face1, db_path='member', enforce_detection=False)

# 1. 取得第一個 DataFrame
df = results[0]

# 2. 檢查 DataFrame 是否為空
if df.shape[0] > 0:  # 使用 shape[0] 檢查列數更精確
    # 診斷用：印出所有欄位看看
    # print(f"目前的欄位有: {df.columns.tolist()}")

    # 尋找包含距離資訊的欄位 (通常包含 'cosine', 'euclidean', 或 'distance')
    dist_cols = [col for col in df.columns if any(keyword in col for keyword in ['cosine', 'euclidean', 'distance'])]

    if dist_cols:
        target_col = dist_cols[0]
        # 進行門檻過濾
        match_df = df[df[target_col] <= 0.25]

        if not match_df.empty:
            identity_path = match_df.iloc[0]['identity']
            file_name = identity_path.split('/')[-1]
            print(f"找到符合的人臉：{file_name} (距離: {match_df.iloc[0][target_col]:.4f})")
        else:
            print(f"雖然資料庫有 7 筆資料，但與目標的距離皆大於 0.25 (最接近的一筆距離為: {df[target_col].min():.4f})")
    else:
        print("錯誤：在回傳結果中找不到距離相關欄位。")
else:
    print("搜尋結束：資料庫中完全沒有符合的人臉 (結果為空)。")
```

## fer：偵測臉部表情(issue)
