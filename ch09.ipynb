接著要做import與Initializing的動作，使『googlenews』這容器代表套件中『GoogleNews()』的方法。

```python
from GoogleNews import GoogleNews
googlenews = GoogleNews()
```

選擇你要搜尋的google news的語言，通常最常用的就是『en , cn , jp…』。在官方文件他是『googlenews = GoogleNews(lang='en')』撰寫，兩者皆可。

```python
googlenews.setlang('cn')
```

開始搜尋並儲存到google套件的暫存區時，每次再重新抓取新的資料前，都要先清空googlenews的陣列，這邊就使用『clear()』。接著我們要設定新聞搜尋的區間日期，如上列程式碼第二段，要注意的是時間填寫的格式『mm/dd/yy』。

```python
googlenews.clear()
googlenews.setTimeRange('03/01/2026','03/10/2026')
```

進入搜尋關鍵字的區塊，假如你要搜尋中文字的話，要注意『encode字串編碼』的地方，也可以在設定容器的時候再『Googlenews(在這邊輸入編碼模式)』。而搜尋的方法請使用『search()』，在括弧中輸入要搜尋的關鍵字。

```python
googlenews.setencode('utf-8')
#googlenews = GoogleNews(encode='utf-8')
googlenews.search('林口')
```

把內容印出來的階段，而請先看到第一個容器，『time』是在設定搜尋的頁數，再來我們透過搜尋的頁數來進行for loop的印出新聞內容。

在for loop中，我們確保安全在清空一次陣列，再來使用『getPage()』取得該頁面所有新聞，由於我們想要把新聞內容都印出來，所以就新增一個儲存新聞內容的容器，這邊我們就以result來儲存，而要取得該頁的結果要使用『googlenews.result()』方法。

而儲存在result中的新聞可能有好幾筆，googlenews套件會把每一則都一筆一筆的放在裡面，所以為了要取得每一筆，我們就在需要使用一次for-loop把每一筆都印出來。最主要印出來的功能就是在『result[哪一筆新聞]』這邊，而下面的程式碼是我想要儲存該新聞連結的陣列『UrlList.append()』將資料塞入陣列中就使用『append()』。

```python
times = 1
UrlList =[]
for i in range(times):
    googlenews.clear()
    googlenews.getpage(i)
    result = googlenews.result()
    for n in range(len(result)):
        print(n)
        print(result[n])
        temp = result[n]
        print(temp['link'])
        UrlList.append(temp['link'])
```
