# 基本自然語言處理

在資訊科學的領域中，「自然語言」（Natural Language, NL）是指人類日常溝通所使用的語言（如中文、英文）。將其引入程式開發中，主要涉及兩個截然不同的層面：開發介面與處理技術。

1. 程式語言的「自然語言化」
早期的程式語言如組合語言，充滿了晦澀難懂的指令。隨著技術演進，程式語言正朝著「人類可讀性」不斷靠攏。例如 Python 或 SQL，其語法結構大量借鑑了英文的邏輯與詞彙，使得非技術人員也能大致理解程式碼的意圖。近年來，隨著 生成式 AI（如大語言模型）的崛起，我們甚至進入了「自然語言程式化」的時代。開發者只需用日常語言描述邏輯，AI 就能將其轉譯為可執行的程式碼，極大地降低了開發門檻。

2. 自然語言處理 (NLP)
這是指程式如何「理解」與「生成」人類語言。這並非單純的字串比對，而是涉及複雜的計算語言學技術，通常包含以下階段：

- 斷詞與標記：將句子拆解為詞彙，並辨識其詞性。
- 語義分析：理解字裡行間的深層含義與意圖。
- 上下文關聯：維持對話的連貫性，如同與真人交談。

## jieba：最常用中文分詞工具

```python
import jieba
```

```python
sentence = '我今天要到台北松山機場出差！'
breakword = jieba.cut(sentence, cut_all=False)
print('精確模式：' + '|'.join(breakword))

breakword = jieba.cut(sentence, cut_all=True)
print('全文模式：' + '|'.join(breakword))

breakword = jieba.cut_for_search(sentence)
print('搜索引擎模式：' + '|'.join(breakword))
```

```python
jieba.set_dictionary('dict.txt.big.txt')
#自訂詞庫
jieba.load_userdict('user_dict_test.txt')
sentence = '這部電影很好看，是我的朋友陳國文主演的。'
breakword = jieba.cut(sentence, cut_all=False)
print('|'.join(breakword))
```

```python
jieba.set_dictionary('dict.txt.big.txt')
jieba.load_userdict('user_dict_test.txt')
#加入停用詞
with open('stopWord_test.txt', 'r', encoding='utf-8-sig') as f:
    stops = f.read().split('\n')
sentence = '這部電影很好看，是我的朋友陳國文主演的。'
breakword = jieba.cut(sentence, cut_all=False)
words = []
for word in breakword:
    if word not in stops:
        words.append(word)
print('|'.join(words))
```

```python
from sumy.parsers.html import HtmlParser #issue
from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer as Summarizer
from sumy.nlp.stemmers import Stemmer
from sumy.utils import get_stop_words
```

```python
LANGUAGE = "chinese"
# LANGUAGE = "english"
SENTENCES_COUNT = 5 #摘要句數
# SENTENCES_COUNT = 10
url = "https://news.ltn.com.tw/news/life/breakingnews/3649202"
# url = "https://en.wikipedia.org/wiki/Automatic_summarization"
parser = HtmlParser.from_url(url, Tokenizer(LANGUAGE))
summarizer = Summarizer(Stemmer(LANGUAGE))
summarizer.stop_words = get_stop_words(LANGUAGE)
sumies = summarizer(parser.document, SENTENCES_COUNT)
for i, sentence in enumerate(sumies):
    print('{}. {}'.format(i+1, sentence))
```

## wordcloud：文字雲

```python
import jieba
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import requests
```

```python
!wget -O font.ttf https://github.com/googlefonts/noto-cjk/raw/main/Sans/OTF/TraditionalChinese/NotoSansCJKtc-Regular.otf
```

```python
# 1. 準備文字資料
text = "人工智慧正在改變世界，Python 是學習大數據與機器學習的最佳工具，文字雲視覺化非常有趣。"

# 2. 使用 jieba 進行斷詞
seg_list = jieba.cut(text, cut_all=False)
seg_text = " ".join(seg_list)

# 3. 設定文字雲參數
wc = WordCloud(
    font_path='font.ttf',     # 設定字體路徑
    width=800,               # 寬度
    height=400,              # 高度
    background_color='white', # 背景顏色
    max_words=100            # 最大顯示單字量
)

# 4. 產生文字雲
wc.generate(seg_text)

# 5. 顯示結果
plt.figure(figsize=(10, 5))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off") # 隱藏座標軸
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import jieba
from collections import Counter
from PIL import Image
import numpy as np
import requests

# 1. 讀取文本
text = open('travel.txt', "r", encoding="utf-8").read()

# 2. 處理停用詞 (增加錯誤檢查防止檔案不存在)
try:
    with open('stopWord_cloudmod.txt', 'r', encoding='utf-8-sig') as f:
        stops = f.read().split('\n')
except FileNotFoundError:
    stops = []
    print("提示：找不到停用詞檔案，將不使用停用詞篩選。")

# 3. 斷詞與統計
terms = [t for t in jieba.cut(text, cut_all=False) if t not in stops and len(t) > 1]
diction = Counter(terms)

# 4. 下載並設定字體
font_url = "https://drive.google.com/uc?id=1QdaqR8Setf4HEulrIW79UEV_Lg_fuoWz&export=download"
font_data = requests.get(font_url).content
with open('taipei_sans_tc_beta.ttf', 'wb') as f:
    f.write(font_data)

# 5. 處理遮罩與生成文字雲
try:
    mask_img = np.array(Image.open("heart.png"))
    wordcloud = WordCloud(background_color="white", mask=mask_img, font_path='taipei_sans_tc_beta.ttf', width=800, height=800)
except:
    print("提示：找不到 heart.png，將生成方形文字雲。")
    wordcloud = WordCloud(background_color="white", font_path='taipei_sans_tc_beta.ttf')

# 重要：生成內容
wordcloud.generate_from_frequencies(frequencies=diction)

# 6. 顯示與存檔
plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()

wordcloud.to_file("bookCloud.png")
```
